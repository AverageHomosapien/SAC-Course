```python
def plot_df(old_df, step):
    new_df = pd.DataFrame()
    for i in range(0, len(old_df), step):
        new_df = new_df.append(old_df[0+i:step+i].mean(), ignore_index=True)
    plt.plot(new_df)
    return new_df
```
